```python
def convert_gender(cell):
    if cell == "male":
        return 1
    else:
        return 0
    
```

```python
import pandas as pd
df = pd.read_csv("csvs/titanic.csv", header=None,skiprows = 1, na_values=["NaN"],converters = {"Sex": convert_gender}, names = ["ID","Survived","Pclass","Name","Sex","Age","Sib","Parch","Ticket","Fare","Cabin","Embarked"])
df
```

```python
df.columns 
```

```python
df.drop(['ID','Name','Parch','Ticket','Cabin','Embarked','Sib'], axis = 1, inplace = True)
df
```

```python
df.dropna(how= "any", inplace = True)
df
```

```python
inputs = df.drop(['Survived'], axis =1)
target = df.Survived
inputs
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
tree = tree.DecisionTreeClassifier
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2, random_state=42)
tree
```
