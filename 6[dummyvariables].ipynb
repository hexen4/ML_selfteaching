```python
import pandas as pd
import numpy as np
df = pd.read_csv("csvs//homeprice.csv")
df
```

```python
dummies = pd.get_dummies(df.town,dtype=float)
dummies
```

```python
df2= pd.concat([df,dummies], axis = 1)
df2.drop(['town','robinsville'], axis = 1, inplace = True)
df2
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
x = df2.drop('price', axis = 1)
y = df2.price
x
```

```python
reg.fit(x,y)
```

```python
w = np.array([2800,0,1])
reg.predict([w])
```

```python
reg.score(x,y)
```
